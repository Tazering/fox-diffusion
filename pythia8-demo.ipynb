```python
import pythia8
```

```python
# create a pythia object
pythia = pythia8.Pythia()
```

```python
# configure pythia
# if the output is true, then we are good
pythia.readString("Beams:eCM = 8000.")
pythia.readString("HardQCD:all = on")
pythia.readString("PhaseSpace:pTHatMin = 20.")
```

```python
# initialize pythia
pythia.init()
```

```python

mult = pythia8.Hist("charged multiplicity", 100, -.05, 788.5)
```

```python
# generate next event
pythia.next()
```

```python
# pythia.event seems to store a bunch of particle objects
for prt in pythia.event:
    print(prt)
```

```python
num_events = 100

# populate mult list with events
for event_itr in range(0, num_events):
    if not pythia.next(): continue # check if there is an error or not

    nCharged = 0

    for prt in pythia.event:
        if prt.isFinal() and prt.isCharged(): nCharged += 1
    
    mult.fill(nCharged)
```

```python
# get statistics and histogram
pythia.stat()
```
